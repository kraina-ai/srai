```python
from srai.datasets import TDrive
import contextily
import matplotlib.pyplot as plt

%load_ext dotenv
%dotenv
```

```python
t_drive = TDrive("../../srai/datasets/config/t_drive.yaml")
```

```python
gdf = t_drive.load()
```

```python
gdf.head()
```

```python
neigh_wm = gdf.iloc[-10:].to_crs(epsg=3857)
ax = neigh_wm.plot(figsize=(10, 8), alpha=0.5, edgecolor="r")
contextily.add_basemap(ax)
plt.show()
```
